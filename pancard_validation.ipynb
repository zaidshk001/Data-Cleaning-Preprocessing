```python
import re

def is_sequential(s):
    return (
        all(ord(s[i]) + 1 == ord(s[i+1]) for i in range(len(s)-1)) or
        all(ord(s[i]) - 1 == ord(s[i+1]) for i in range(len(s)-1))
    ) # this checks if the characters are in sequential order either increasing or decreasing by comparing the ASCII values of the characters

def has_adjacent_duplicates(s):
    return any(s[i] == s[i+1] for i in range(len(s)-1)) # this checks if there are any adjacent characters that are the same

def validate_pan(pan):
    # Basic format check
    if not re.match(r'^[A-Z]{5}[0-9]{4}[A-Z]$', pan): # this regex checks if the PAN number has 5 uppercase letters followed by 4 digits and then 1 uppercase letter, why did we use not match? because we want to return false if the format is not correct
        return False
    
    first_part = pan[:5]
    digits_part = pan[5:9]
    
    # Alphabet checks
    if has_adjacent_duplicates(first_part):
        return False
    if is_sequential(first_part):
        return False
    
    # Digit checks
    if has_adjacent_duplicates(digits_part):
        return False
    if is_sequential(digits_part):
        return False
    
    return True
```

```python
df['PAN_Category'] = df['Pan_Numbers'].apply(
    lambda x: "Valid PAN" if validate_pan(x) else "Invalid PAN"
)
```
