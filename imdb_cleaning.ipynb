```python
from dateutil import parser
import pandas as pd

def robust_date_parser(x):
    if pd.isna(x):
        return pd.NaT
    
    try:
        return parser.parse(str(x), dayfirst=False)
    except:
        return pd.NaT
```

```python
def find_outliers(df, col):
    
    q1, q3 = np.percentile(df[col].dropna(), [25, 75])
    
    iqr = q3 - q1
    
    lw = q1 - 1.5 * iqr
    uw = q3 + 1.5 * iqr
    
    outliers = df[(df[col] < lw) | (df[col] > uw)]
    
    return outliers
```
