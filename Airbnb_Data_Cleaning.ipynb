```python
columns_to_keep = [ 'NAME', 'host id', 'host_identity_verified', 'host name',
       'neighbourhood group', 'neighbourhood', 'lat', 'long', 'country',
       'country code', 'instant_bookable', 'cancellation_policy', 'room type',
       'Construction year', 'price', 'service fee', 'minimum nights',
       'number of reviews', 'last review']

columns_to_drop = [col for col in df.columns if col not in columns_to_keep]
```

```python
print(len(columns_to_keep))
print(len(columns_to_drop))
```

```python
# now filter the dataframe to keep only the columns we want to keep and drop the rest.
df = df[columns_to_keep]
```

```python
df.shape
```

```python
df.rename(columns = {"NAME": "name", "Construction year": "construction_year"} , inplace = True)

df.columns
```

```python
# if you want to change the entire column header to title case you can do it like this.

df.columns = df.columns.str.title()
```

```python
# method 2 

for col in df.columns:
    df.rename(columns = {col: col.title().replace(" ", "_")}, inplace = True)
```

```python
df.columns
```

```python
# method 3 using empty list and for loop

new_columns = []

for col in df.columns:
    new_col = col.title().replace(" ", "_")
    new_columns.append(new_col)

```

```python
df_new = df.copy() # we create a copy of the original dataframe to avoid modifying it directly
df_new.columns = new_columns # we assign the new column names to the copy of the dataframe
df_new.columns
```

```python
df.duplicated().sum()
```

```python
df[df.duplicated(keep= 'first')] # what keep = false does is it shows all the duplicated rows instead of just one of them.
# keep = first shows all the duplicated rows except the first occurrence of the duplicate, so first occurence will be considered as unique and the rest will be considered as duplicates. 
# yaane first row is false and the rest of the duplicates will be true.
# while keep = last shows all the duplicated rows except the last occurrence of the duplicate. all the duplicates will be true except the last one which will be false.
# keep = false shows all the duplicated rows including the first and last occurrence of the duplicate. all the duplicates will be true. so it will show all the duplicated rows.
```

```python
df[df.duplicated(keep=False)].sort_values(by=df.columns.tolist()) # this will show us the duplicated rows sorted by all the columns in the dataframe.
```

```python
# now we want to remove the duplicate and keep only one of the duplicates, we can use the drop_duplicates() method.

df.drop_duplicates(inplace=True) # so this rows will drop all the duplicate rows and keep only one of the duplicates. 
# in this case it will keep the first occurrence of the duplicate and drop the rest. and the duplicate is  when all the columns are similar to the other row.
```

```python
# I want to fill the missing host name with the value of host id because it is a unique identifier for each host and it will not affect our analysis as much as filling it with a common value like "unknown" or "missing". so we will fill the missing values in the host name column with the corresponding values in the host id column.

df.groupby('Host_Id')['Host_Name'].nunique() > 1
```

```python
df['Host_Name'] = df.groupby('Host_Id')['Host_Name'].transform(lambda x: x.fillna(method='ffill').fillna(method='bfill'))
```
