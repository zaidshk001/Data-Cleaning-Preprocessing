```python
from word2number import w2n
import re

def clean_salary(x):
    if pd.isna(x):
        return None
    
    x = str(x).lower().strip() # Convert to string, lowercase, and remove whitespace
    
    # Case 1: already numeric
    if x.isdigit(): # If the string is purely digits, convert to int
        return int(x)
    
    # Case 2: contains 'k' (e.g., 75k)
    if "k" in x:
        return float(x.replace("k", "")) * 1000
    
    # Case 3: text numbers (e.g., "sixty thousand")
    try:
        return w2n.word_to_num(x)
    except: # If conversion fails, we will try regex next
        pass
    
    # Case 4: extract numbers using regex
    nums = re.findall(r"\d+", x) # Find all sequences of digits in the string and return them as a list
    if nums:
        return int(nums[0]) # 
    
    return None
```

```python
df['Salary'] = df["Salary"].apply(clean_salary) # apply the cleaning function to the Salary column
```

```python
# the salary column looks fairly uniform with no major outliers and the skewness and kurtosis values are close to 0, so we can impute null values with mean or median. We will use median for Salary column.
# also we will try to impute null values based on median salary for a position

df['Salary'] = df.groupby(['Department','Position'])['Salary']\
    .transform(lambda x: x.fillna(x.median())) # what transform does is it applies the lambda function to each group of 'Department' and Position' and fills null values in 'Salary' with the median salary of that position group.
# is it not possible to do it without using transform? like using apply or something else? Yes, we can also use apply to achieve the same result. Here's how you can do it with apply:
# df['Salary'] = df.groupby('Position')['Salary'].apply(lambda x: x.fillna(x.median()))
# we are using transform because it retains the shape of the original DataFrame, while apply would return a Series with a different index, which might not align properly with the original DataFrame when we try to assign it back to the 'Salary' column. Transform is specifically designed for this kind of operation where we want to fill missing values based on group-level statistics while keeping the original DataFrame structure intact.
```

```python
from dateutil import parser
import pandas as pd

def robust_date_parser(x):
    if pd.isna(x):
        return pd.NaT
    
    try:
        return parser.parse(str(x), dayfirst=False)
    except:
        return pd.NaT
```

```python
df['Joining Date_clean'] = df['Joining Date'].apply(robust_date_parser)
```
